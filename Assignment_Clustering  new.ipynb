```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
X=df.iloc[:,1:]
X
y=df['Type']
```

```python
pca=PCA(n_components=3)
pca_component=pca.fit_transform(X)
pca_component
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hierarchical_clustering=AgglomerativeClustering(affinity='euclidean')
```

```python
model1=hierarchical_clustering.fit(pca_component)
```

```python
Pred_hc=model1.fit_predict(pca_component)
Pred_hc
```

```python
from scipy.cluster.hierarchy import dendrogram , ward
linked_array=ward(pca_component)
```

```python
dendrogram(linked_array,truncate_mode='lastp',p=30,leaf_font_size=12.,leaf_rotation=90.,show_contracted=True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
accuracy=silhouette_score(pca_component,Pred_hc)
accuracy
```

```python
kmeans=KMeans()
kmeans.fit(X)
```

```python
pred_X=kmeans.fit_predict(X)
pred_X
```

```python
accuracy_X=silhouette_score(X,pred_X)
accuracy_X
```

```python
wcss=[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(pca_component)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
