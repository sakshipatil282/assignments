# 1. Data Cleaning and Preparation

Load the dataset into a DataFrame or equivalent data structure.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Cardiotocographic.csv")
df
```

Handle missing values appropriately (e.g., imputation, deletion).

```python
missing_values = df.isnull().sum()
print(f"Missing values in each column:\n{missing_values}")
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df_imputed = df.fillna(df.mean())
```

```python
missing_values_after_imputation = df_imputed.isnull().sum()

missing_values_after_imputation
```

Identify and correct any inconsistencies in data types (e.g., numerical values stored as strings).

```python
df.dtypes
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

# 2. Statistical Summary

Provide a statistical summary for each variable in the dataset, including measures of central tendency (mean, median) and dispersion (standard deviation, interquartile range).

```python
summary = df.describe()
print("Statistical Summary:\n", summary)
```

```python
mean_values = df.mean()
median_values = df.median()
std_values = df.std()
```

```python
print(f"Means:\n{mean_values}")
print(f"Medians:\n{median_values}")
print(f"Standard Deviations:\n{std_values}")
```

```python
iqr_values = df.quantile(0.75) - df.quantile(0.25)
print(f"IQR Values:\n{iqr_values}")
```

Highlight any interesting findings from this summary.

```python
print("Interesting Findings:\n")
for col in df.columns:
    skewness = df[col].skew()
    if skewness > 1 or skewness < -1:
        print(f"{col} has high skewness: {skewness}")
```

# 3. Data Visualization

Create histograms or boxplots to visualize the distributions of various numerical variables.

```python
sns.set(style='whitegrid')
```

```python
df.hist(bins=30, figsize=(12, 8), edgecolor='black')
plt.suptitle('Histograms of Numerical Variables')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title('Boxplot of Numerical Variables')
plt.show()
```

Use bar charts or pie charts to display the frequency of categories for categorical variables.


```python
plt.figure(figsize=(10, 6))
df['Width'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Bar Chart of Categorical Variable')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

Generate scatter plots or correlation heatmaps to explore relationships between pairs of variables.

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='LB', y='AC', data=df_imputed)
plt.title('Scatter Plot of LB vs AC')
plt.show()

strong_correlations = correlation_matrix[(correlation_matrix > 0.7) | (correlation_matrix < -0.7)]
print(strong_correlations)
```

Employ advanced visualization techniques like pair plots, or violin plots for deeper insights.

```python
sns.pairplot(df)
plt.suptitle('Pair Plot of Variables', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, inner='quartile')
plt.title('Violin Plot of Numerical Variables')
plt.show()
```
