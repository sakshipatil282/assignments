# Choose an appropriate distance metric and value for K

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
k_values = range(1,3)
accuracy_scores = [19,57]
```

```python
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')
    accuracy_scores.append(scores.mean())
```

```python
print(len(k_values))
print(len(accuracy_scores))
```

```python
plt.plot(k_values, accuracy_scores)
plt.xlabel('k')
plt.ylabel('Cross-Validated Accuracy')
plt.title('K value vs. Accuracy')
plt.show()
```

```python
best_k = k_values[accuracy_scores.index(max(accuracy_scores))]
print(f'Best k: {best_k}')
```

# Evaluate the classifier

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
```

```python
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```

# Visualize the decision boundaries

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train_pca, y_train)
```

```python
import numpy as np
```

```python
x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
```

```python
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, edgecolors='k', marker='o')
plt.title('Decision Boundaries with KNN')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
