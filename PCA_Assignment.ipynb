# Task 2: Dimensionality Reduction with PCA

Standardize the Features

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

 Implement PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(scaled_data)
```

Determine the Optimal Number of Principal Components

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_, marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
cumulative_variance = pca.explained_variance_ratio_.cumsum()
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance Explained')
plt.show()
```

Transform the Original Dataset

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)
```

```python
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
```

# Task 3: Clustering with Original Data

Apply K-Means Clustering

```python
from sklearn.cluster import KMeans
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)
df['Cluster'] = clusters
```

Visualize Clustering Results

```python
print(principal_df.columns)
```

```python
principal_df['Cluster'] = clusters
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=principal_df)
plt.title('K-Means Clustering with Original Data')
plt.show()
```

 Evaluate Clustering Performance

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(scaled_data, clusters)
print('Silhouette Score:', score)
```

```python
from sklearn.metrics import davies_bouldin_score
db_index = davies_bouldin_score(scaled_data, clusters)
print('Davies–Bouldin Index:', db_index)
```

# Task 4: Clustering with PCA Data

Apply K-Means Clustering to PCA-Transformed Data

```python
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(principal_components)
principal_df['Cluster'] = clusters_pca
```

 Visualize Clustering Results

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=principal_df)
plt.title('K-Means Clustering with PCA Data')
plt.show()
```

Evaluate Clustering Performance

```python
score_pca = silhouette_score(principal_components, clusters_pca)
print('Silhouette Score (PCA Data):', score_pca)
```

```python
db_index_pca = davies_bouldin_score(principal_components, clusters_pca)
print('Davies–Bouldin Index (PCA Data):', db_index_pca)
```
